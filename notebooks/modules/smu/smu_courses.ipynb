```python
OUTPUT_CSV = "smu_courses.csv"
OUTPUT_JSON = "smu_courses.json"
SLEEP_SECONDS = 1

HEADERS = {
    "User-Agent": "Mozilla/5.0"
}

def extract_code(url):
    m = re.search(r"/courses/(\d+)", url)
    return m.group(1) if m else ""

def clean_text(text):
    if not text:
        return ""
    return re.sub(r"\s+", " ", str(text)).strip()

def is_bad_text(text):
    t = text.lower()
    bad_patterns = [
        "all rights reserved",
        "copyright",
        "©",
        "singapore management university",
    ]
    return any(x in t for x in bad_patterns)

def extract_description(html):
    soup = BeautifulSoup(html, "html.parser")

    # extract from "Description"
    label = soup.find(
        lambda tag: tag.name in ["h1", "h2", "h3", "h4", "strong", "b", "div", "span", "td"]
        and clean_text(tag.get_text()).lower() == "description"
    )

    if label:
        for nxt in label.find_all_next(["p", "div", "td", "span", "li"], limit=25):
            txt = clean_text(nxt.get_text(" ", strip=True))
            if len(txt) > 80 and not is_bad_text(txt) and txt.lower() != "description":
                return txt

    candidates = []
    for tag in soup.find_all(["p", "div", "td", "span", "li"]):
        txt = clean_text(tag.get_text(" ", strip=True))
        if len(txt) > 120 and not is_bad_text(txt):
            candidates.append(txt)

    keywords = [
        "this course", "students", "introduces", "examines", "provides",
        "focuses", "covers", "develop", "designed", "learn", "understand"
    ]

    scored = []
    for txt in candidates:
        score = sum(1 for kw in keywords if kw in txt.lower())
        scored.append((score, len(txt), txt))

    if scored:
        scored.sort(reverse=True)
        return scored[0][2]

    return ""

def scrape_one(course_name, course_type, url):
    code = extract_code(url)

    try:
        r = requests.get(url, headers=HEADERS, timeout=30)
        r.raise_for_status()

        description = extract_description(r.text)

        return {
            "course": course_name,
            "type": course_type,
            "code": code,
            "description": description,
            "url": url,
            "status": "ok" if description else "no_description"
        }

    except Exception as e:
        return {
            "course": course_name,
            "type": course_type,
            "code": code,
            "description": "",
            "url": url,
            "status": f"error: {e}"
        }


all_rows = []

for course_name, type_groups in course_groups.items():
    for course_type, urls in type_groups.items():
        print(f"\n--- Scraping {course_name} | {course_type} ({len(urls)} URLs) ---")

        for url in urls:
            print("Scraping:", url)
            row = scrape_one(course_name, course_type, url)
            all_rows.append(row)
            time.sleep(SLEEP_SECONDS)

df_new = pd.DataFrame(all_rows)


df_final = df_new[["course", "type", "code", "description"]].copy()
df_final.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")

print(f"\nSaved to {OUTPUT_CSV}")
df_final
```
