```python
import json
import pandas as pd

# uploaded file path
file_path = "../../data/courses data 2/Accountancy.json"

# elective codes you want to mark
accountancy_electives = {
    "ACC2711", "ACC2712", "ACC3711", "ACC3713", "ACC3714", "ACC3715",
    "ACC3717", "ACC4711", "ACC4712", "ACC4713", "ACC4715", "ACC4751",
    "ACC4761", "ACC4761A"
}

with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

degree = data.get("Programme", "").strip()
codes = data.get("Course Codes", [])

# handle either a list or a comma-separated string
if isinstance(codes, str):
    codes = [c.strip() for c in codes.split(",") if c.strip()]
elif isinstance(codes, list):
    cleaned = []
    for c in codes:
        if isinstance(c, str):
            cleaned.extend([x.strip() for x in c.split(",") if x.strip()])
    codes = cleaned

rows = []
for code in codes:
    course_type = "elective" if code in accountancy_electives else "core"
    rows.append({
        "course": "accountancy",
        "type": course_type,
        "code": code,
        "description": ""
    })

df = pd.DataFrame(rows)

# dedupe just in case
df = df.drop_duplicates(subset=["course", "type", "code"]).reset_index(drop=True)

# save
df.to_csv("nus_courses.csv", index=False, encoding="utf-8-sig")

df
```

```python
def get_level(code):
    try:
        for char in str(code):
            if char.isdigit():
                return int(char) * 1000
    except:
        return None

def depth_weight(code):
    level = get_level(code)
    return {
        1000: 1.0,
        2000: 1.2,
        3000: 1.5,
        4000: 1.8
    }.get(level, 1.0)

def type_weight(t):
    t = str(t).lower()
    if "core" in t:
        return 1.0
    elif "elective" in t:
        return 0.7
    return 0.8

df_final["weight"] = df_final.apply(
    lambda row: (
        depth_weight(row["code"]) *
        type_weight(row["type"])
    ),
    axis=1
)

df_final.to_csv(master_csv, index=False, encoding="utf-8-sig")
df
```
